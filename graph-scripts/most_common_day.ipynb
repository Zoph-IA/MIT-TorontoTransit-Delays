```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data_updated.csv')
```

```python
day_counts = data['Day'].value_counts()
```

```python
mean_value = day_counts.mean()
min_value = day_counts.min()
max_value = day_counts.max()
```

```python
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink']
```

```python
plt.figure(figsize=(10, 6))
day_counts.plot(kind='bar', color=colors, label='Occurrences')
plt.axhline(mean_value, color='red', linestyle='--', label=f'Mean: {mean_value:.0f}')
plt.axhline(min_value, color='green', linestyle='--', label=f'Min: {min_value:.0f}')
plt.axhline(max_value, color='blue', linestyle='--', label=f'Max: {max_value:.0f}')

for i, count in enumerate(day_counts):
    plt.text(i, count + 5, str(count), ha='center', fontsize=10)

plt.title('Most Common Day of the Week', fontsize=16)
plt.xlabel('Day', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.savefig('plots/most_common_day.png')
plt.show()
plt.close()
```
