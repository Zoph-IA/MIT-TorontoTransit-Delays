```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data_updated.csv')
```

```python
plt.figure(figsize=(10, 6))
location_counts = df['Location'].value_counts().head(10)
colors = plt.cm.tab10(range(len(location_counts)))

bars = plt.bar(location_counts.index, location_counts.values, color=colors)

for bar, count in zip(bars, location_counts.values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 5, str(count), ha='center', fontsize=10)

plt.title('Top 10 Locations with Most Incidents', fontsize=16)
plt.xlabel('Location', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

plt.savefig('plots/top_locations_incidents.png')
plt.show()
plt.close()
```
