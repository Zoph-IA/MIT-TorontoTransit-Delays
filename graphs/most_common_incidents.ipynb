```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data_updated.csv')
```

```python
plt.figure(figsize=(10, 6))
incident_counts = df['Incident'].value_counts()
incident_counts.plot(kind='bar', color=['skyblue', 'orange', 'green', 'red', 'purple', 'pink', 'gray'])
plt.title('Most Common Incident Types', fontsize=16)
plt.xlabel('Incident Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('plots/most_common_incidents.png')
plt.show()
plt.close()
```
