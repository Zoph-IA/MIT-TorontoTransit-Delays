```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data_updated.csv')
```

```python
overall_mean = df['Line'].value_counts().mean()
overall_min = df['Line'].value_counts().min()
overall_max = df['Line'].value_counts().max()
```

```python
plt.figure(figsize=(10, 6))
line_counts = df['Line'].value_counts().head(10)
colors = plt.cm.tab10(range(len(line_counts)))

bars = plt.bar(line_counts.index, line_counts.values, color=colors, alpha=0.8, label='Occurrences')

plt.axhline(overall_mean, color='red', linestyle='--', label=f'Mean: {overall_mean:.0f}')
plt.axhline(overall_min, color='green', linestyle='--', label=f'Min: {overall_min:.0f}')
plt.axhline(overall_max, color='blue', linestyle='--', label=f'Max: {overall_max:.0f}')

for bar, count in zip(bars, line_counts.values):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 5, str(count), ha='center', fontsize=10)

plt.title('Top 10 Lines with Most Incidents', fontsize=16)
plt.xlabel('Line', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

plt.savefig('plots/top_lines_incidents.png')
plt.show()
plt.close()
```
